```python
import numpy as np
import pandas as pd
```

```python
file_paths = [
    "/content/MW-NIFTY-AUTO-26-Oct-2024.csv",
    "/content/MW-NIFTY-BANK-26-Oct-2024.csv",
    "/content/MW-NIFTY-COMMODITIES-26-Oct-2024.csv",
    "/content/MW-NIFTY-CONSUMER-DURABLES-26-Oct-2024.csv",
    "/content/MW-NIFTY-ENERGY-26-Oct-2024.csv",
    "/content/MW-NIFTY-FINANCIAL-SERVICES-25_50-26-Oct-2024.csv",
    "/content/MW-NIFTY-FINANCIAL-SERVICES-26-Oct-2024.csv",
    "/content/MW-NIFTY-FMCG-26-Oct-2024.csv",
    "/content/MW-NIFTY-HEALTHCARE-INDEX-26-Oct-2024.csv",
    "/content/MW-NIFTY-IT-26-Oct-2024.csv",
    "/content/MW-NIFTY-INDIA-CONSUMPTION-26-Oct-2024.csv",
    "/content/MW-NIFTY-MEDIA-26-Oct-2024.csv",
    "/content/MW-NIFTY-METAL-26-Oct-2024.csv",
    "/content/MW-NIFTY-OIL-&-GAS-26-Oct-2024.csv",
    "/content/MW-NIFTY-REALTY-26-Oct-2024.csv",
    "/content/MW-NIFTY-PRIVATE-BANK-26-Oct-2024.csv",
    "/content/MW-NIFTY-MIDSMALL-HEALTHCARE-26-Oct-2024.csv",
    "/content/MW-NIFTY-PSU-BANK-26-Oct-2024.csv"
]
```

```python
dataframes = []

for i, path in enumerate(file_paths, start=1):

    try:
        df = pd.read_csv(path)
        df.columns = [col.strip() for col in df.columns]  # Strip whitespace/newline from column names
        df = df[['SYMBOL']][1:]  # Retain only the SYMBOL column
        dataframes.append(df)  # Append the processed DataFrame to the list

    except FileNotFoundError:
        print(f"File not found: {path}")
```

```python
stock_symbols = pd.DataFrame(columns=["SYMBOL"])

for i, df in enumerate(dataframes):
    for symbol in df["SYMBOL"].values:  # Access the values in the SYMBOL column
        if symbol not in stock_symbols["SYMBOL"].values:
            stock_symbols = pd.concat([stock_symbols, pd.DataFrame({"SYMBOL": [symbol]})], ignore_index=True)
```
