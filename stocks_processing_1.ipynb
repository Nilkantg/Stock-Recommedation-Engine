```python
dataframes = []
for i in range(1, 18):
  # Access the correct dataframe using globals()
  current_df = globals()[f"df{i}"]
  # Strip the column names of the current dataframe
  current_df.columns = [col.strip() for col in current_df.columns]
  # Append the 'SYMBOL' column to the list
  dataframes.append(current_df["SYMBOL"][1:])

# Now dataframes is a list containing the 'SYMBOL' column from each dataframe
```

```python
dataframes[1]
```

```python
stock_symbols = pd.DataFrame(columns=["SYMBOL"])

for i, df in enumerate(dataframes):

  for symbol in df:
    if symbol not in stock_symbols["SYMBOL"].values:

      stock_symbols = pd.concat([stock_symbols, pd.DataFrame({"SYMBOL": [symbol]})], ignore_index=True)
```
