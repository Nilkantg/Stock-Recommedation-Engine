```python
dataset = pd.DataFrame(columns=['SYMBOL',
                                'Industry',
                                'Sector',
                                'fiftyTwoWeekHigh',
                                'marketCap',
                                'dividend Rate',
                                'trailing PE',
                                'forward PE',
                                'priceToSales',
                                'Profit Margins',
                                'Operating Margins',
                                'Revenue Growth',
                                'Earnings Growth',
                                'ROA',
                                'ROE',
                                'Volume',
                                'DebtToEquity'])
```

```python

for i, ticker in enumerate(stock_symbols['SYMBOL'].values):

  stock = yf.Ticker(f"{ticker}.NS")
  stock_info = stock.info
  industry = stock_info.get('industry')
  sector = stock_info.get('sector')
  fiftyTwoWeekHigh = stock_info.get('fiftyTwoWeekHigh')
  marketCap = stock_info.get('marketCap')
  dividendRate = stock_info.get('dividendRate')
  trailingPE = stock_info.get('trailingPE')
  forwardPE = stock_info.get('forwardPE')
  priceToSales = stock_info.get('priceToSalesTrailing12Months')
  profitMargins = stock_info.get('profitMargins')
  operatingMargins = stock_info.get('operatingMargins')
  revenueGrowth = stock_info.get('revenueGrowth')
  earningsGrowth = stock_info.get('earningsGrowth')
  ROA = stock_info.get('returnOnAssets')
  ROE = stock_info.get('returnOnEquity')
  volume = stock_info.get('volume')
  debtToEquity = stock_info.get('debtToEquity')

  dataset.loc[i] = [ticker,
                    industry,
                    sector,
                    fiftyTwoWeekHigh,
                    marketCap,
                    dividendRate,
                    trailingPE,
                    forwardPE,
                    priceToSales,
                    profitMargins,
                    operatingMargins,
                    revenueGrowth,
                    earningsGrowth,
                    ROA,
                    ROE,
                    volume,
                    debtToEquity]
```
